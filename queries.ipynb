```python
# Import dependencies.
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load cleaned stroke dataset
file_path = "Resources/stroke_cleaned.csv"
stroke_df = pd.read_csv(file_path)

# Change Hypertension values from 0/1 to strings
stroke_df['Hypertension'] = stroke_df['Hypertension'].replace([0,1],['no hypertension','hypertension'])
# Change Heart Disease values from 0/1 to strings
stroke_df['HeartDisease'] = stroke_df['HeartDisease'].replace([0,1],['no heart disease','heart disease'])
# Change Stroke values from 0/1 to strings
stroke_df['Stroke'] = stroke_df['Stroke'].replace([0,1],['stroke','no stroke'])

stroke_df
```

```python
# Create DataFrame of people with strokes
stroke_yes = stroke_df[stroke_df.Stroke == "stroke"]

stroke_yes
```

```python
# Create DataFrame of people without strokes
stroke_no = stroke_df[stroke_df.Stroke == "no stroke"]

stroke_no
```
