```python
# Import dependencies.
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Load cleaned stroke dataset
file_path = "Resources/stroke_cleaned.csv"
stroke_df = pd.read_csv(file_path)
stroke_df.head(10)
```

```python
# Change Hypertension values from 0/1 to strings
stroke_df['Hypertension'] = stroke_df['Hypertension'].replace([0,1],['no hypertension','hypertension'])

stroke_df.head(10)
```

```python
# Change Heart Disease values from 0/1 to strings
stroke_df['HeartDisease'] = stroke_df['Hypertension'].replace([0,1],['no heart disease','heart disease'])

stroke_df.head(10)
```

```python
# Change Stroke values from 0/1 to strings
stroke_df['Stroke'] = stroke_df['Stroke'].replace([0,1],['no stroke','stroke'])

stroke_df.head(10)
```

```python
# Pie chart showing Gender breakdown
colors = ["magenta", "lightblue"]
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(stroke_df["Gender"].value_counts(),
    labels=stroke_df["Gender"].value_counts(),
    autopct='%.1f%%',
    colors=colors,
    shadow=True,
    startangle=90,
    counterclock=False)
plt.legend(['Female','Male'], loc="best")
plt.title("Stroke Dataset Gender Breakdown", bbox={'facecolor':'0.8', 'pad':5})

plt.show

# Save figure
plt.savefig('Resources/demo_gender.png')
```

```python
# Distribution plot showing Age breakdown
sns.displot(stroke_df["Age"], bins=16).set(title='Stroke Dataset Age Breakdown')

# Save figure
plt.savefig('Resources/demo_age.png')
```

```python
# Pie chart showing Hypertension breakdown
colors = ["grey", "red"]
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(stroke_df["Hypertension"].value_counts(),
    labels=stroke_df["Hypertension"].value_counts(),
    autopct='%.1f%%',
    colors=colors,
    shadow=True,
    startangle=90,
    counterclock=False)
plt.legend(['No Hypertension','Hypertension'], loc="best")
plt.title("Stroke Dataset Hypertension Breakdown", bbox={'facecolor':'0.8', 'pad':5})

plt.show

# Save figure
plt.savefig('Resources/demo_hypertension.png')
```
